```python
# Plotting Numerical Features w.r.t the Target Variable (label)

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

def plot_numerical_hist(df, num_feature_list):

    
    for feature in num_feature_list:
        
        plt.figure(figsize=(14, 6))
        
        # Subplot 1: Not Fraud
        plt.subplot(1, 2, 1)
        sns.histplot(
            data=df[df['isFraud'] == 0],
            x=feature,
            bins=30,
            color='blue',
            label='Not Fraud',
            kde=False
        )
        plt.title(f'{feature} - Not Fraud', fontsize=16)
        plt.xlabel(feature, fontsize=14)
        plt.ylabel('Count', fontsize=14)
        plt.legend()
        
        # Subplot 2: Fraud
        plt.subplot(1, 2, 2)
        sns.histplot(
            data=df[df['isFraud'] == 1],
            x=feature,
            bins=30,
            color='red',
            label='Fraud',
            kde=False
        )
        plt.title(f'{feature} - Fraud', fontsize=16)
        plt.xlabel(feature, fontsize=14)
        plt.ylabel('Count', fontsize=14)
        plt.legend()
        
        plt.tight_layout()
        
        plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Known function for combinations https://stackoverflow.com/questions/464864/get-all-possible-2n-combinations-of-a-list-s-elements-of-any-length

def get_combinations(arr, comb_len):
        if comb_len == 0:
            return [[]]
        combinations = []
        for i in range(len(arr)):
            for next_comb in get_combinations(arr[i+1:], comb_len-1):
                combinations.append([arr[i]] + next_comb)
        return combinations

def plot_cols_to_targetlabel(df, cols_list, max_comb=3):

    df['isFraud'] = df['isFraud'].astype('category')

    for k in range(1, min(len(cols_list), max_comb) + 1):
        comb_list = get_combinations(cols_list, k)
        num_plots = len(comb_list)

        cols_per_row = 3
        rows = (num_plots + cols_per_row - 1) // cols_per_row

        fig, axes = plt.subplots(rows, cols_per_row, figsize=(6 * cols_per_row, 6 * rows))
        axes = axes.flatten() if num_plots > 1 else [axes]

        for idx, comb in enumerate(comb_list):
            ax = axes[idx]

            combination_series = df[comb].astype(str).agg(', '.join, axis=1)
            
            sns.histplot(
                x=combination_series,
                hue=df['isFraud'],
                palette='colorblind',
                ax=ax,
                stat='percent',
                multiple='dodge'
            )

            comb_label = ', '.join(comb)
            ax.set_title(f"Fraud Percentage by {comb_label}", fontsize=14)
            ax.set_ylabel('Percentage (%)', fontsize=12)
            ax.set_xlabel(comb_label, fontsize=12)

            ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

            ax.set_ylim(0, 100)

            for container in ax.containers:
                ax.bar_label(container, fmt='%.1f%%', fontsize=8, label_type='edge')

        for idx in range(num_plots, len(axes)):
            fig.delaxes(axes[idx])

        plt.tight_layout()
        plt.show()

        if k == len(cols_list)-1:
            all_cols_1_or_0 = (df[cols_list] == 1).all(axis=1).replace({True: 'all cols = 1', False: 'all cols = 0'})

            plt.figure(figsize=(8, 6))
            sns.histplot(
                x=all_cols_1_or_0,
                hue=df['isFraud'],
                palette='colorblind',
                stat='percent',
                multiple='dodge'
            )
            plt.title('Fraud Percentage by All Columns = 1 or 0', fontsize=14)
            plt.ylabel('Percentage (%)', fontsize=12)
            plt.xlabel('All Columns Status', fontsize=12)

            plt.ylim(0, 100)

            ax = plt.gca()
            for container in ax.containers:
                ax.bar_label(container, fmt='%.2f%%', fontsize=8, label_type='edge')

            plt.tight_layout()
            plt.show()
```

```python
import pandas as pd
import numpy as np

def cartesian_product(lists):
    result = [[]]
    for pool in lists:
        result = [x + [y] for x in result for y in pool]
    return result


def get_column_combinations(arr, comb_len):
        combinations = []
        for i in range(len(arr)):
            for next_comb in get_combinations(arr[i+1:], comb_len-1):
                combinations.append([arr[i]] + next_comb)
        return combinations


def stepwise_fraud_finding(df, label_name, columns_list):

    train_df = df.dropna(subset=[label_name])

    uniform_subsets = []

    for col in columns_list:
        for value in [0, 1]:
            subset = train_df[train_df[col] == value]
            label_counts = subset[label_name].value_counts()
            total = label_counts.sum()
            if total == 0:
                continue
            max_label_value = label_counts.idxmax()
            max_count = label_counts[max_label_value]
            ratio = max_count / total
            uniform_subsets.append({
                'columns': [(col, value)],
                'label_value': max_label_value,
                'ratio': ratio,
                'count': total
            })

    # Step 3: Check for uniform label values in all combinations of columns
    n_cols = len(columns_list)
    for combination_size in range(2, n_cols + 1):
        # Generate combinations of columns of size combination_size
        combinations_list = get_column_combinations(columns_list, combination_size)
        for combination in combinations_list:
            # Generate all combinations of values (0 or 1) for these columns
            values_list = [[0, 1]] * combination_size
            value_combinations = cartesian_product(values_list)
            for value_combination in value_combinations:
                # Build mask to select subset
                mask = train_df[combination[0]] == value_combination[0]
                for col, val in zip(combination[1:], value_combination[1:]):
                    mask &= (train_df[col] == val)
                subset = train_df[mask]
                if subset.empty:
                    continue
                # Get label_name counts in the subset
                label_counts = subset[label_name].value_counts()
                total = label_counts.sum()
                if total == 0:
                    continue
                max_label_value = label_counts.idxmax()
                max_count = label_counts[max_label_value]
                ratio = max_count / total
                uniform_subsets.append({
                    'columns': list(zip(combination, value_combination)),
                    'label_value': max_label_value,
                    'ratio': ratio,
                    'count': total
                })

    # Sort uniform_subsets by ratio in descending order
    uniform_subsets.sort(key=lambda x: x['ratio'], reverse=True)

    # Initialize the DataFrame to fill and a mask for NAs in the label column
    df_filled = df.copy()
    df_filled[label_name + '_filled'] = df_filled[label_name]
    na_mask = df_filled[label_name].isna()

    # Step 4: Fill NAs with the uniform value in label_name for each subset where ratio == 1
    for subset_info in uniform_subsets:
        if subset_info['ratio'] < 1.0:
            continue  # Only consider subsets with ratio == 1.0 in this step
        cols_values = subset_info['columns']
        label_value = subset_info['label_value']
        # Build mask for subset
        subset_mask = na_mask.copy()
        for col, val in cols_values:
            subset_mask &= (df_filled[col] == val)
        # Fill NAs in label_name_filled
        df_filled.loc[subset_mask, label_name + '_filled'] = label_value
        # Update na_mask
        na_mask = df_filled[label_name + '_filled'].isna()

    # Step 5: Fill NAs with the uniform value in label_name for subsets where ratio >= 0.95
    for subset_info in uniform_subsets:
        if subset_info['ratio'] < 0.95 or subset_info['ratio'] >= 1.0:
            continue  # Skip subsets already processed or with ratio less than 0.95
        cols_values = subset_info['columns']
        label_value = subset_info['label_value']
        # Build mask for subset
        subset_mask = na_mask.copy()
        for col, val in cols_values:
            subset_mask &= (df_filled[col] == val)
        # Fill NAs in label_name_filled
        df_filled.loc[subset_mask, label_name + '_filled'] = label_value
        # Update na_mask
        na_mask = df_filled[label_name + '_filled'].isna()
    
    for subset_info in uniform_subsets:
        if subset_info['ratio'] < 0.92 or subset_info['ratio'] >= .95:
            continue  # Skip subsets already processed or with ratio less than 0.95
        cols_values = subset_info['columns']
        label_value = subset_info['label_value']
        # Build mask for subset
        subset_mask = na_mask.copy()
        for col, val in cols_values:
            subset_mask &= (df_filled[col] == val)
        # Fill NAs in label_name_filled
        df_filled.loc[subset_mask, label_name + '_filled'] = label_value
        # Update na_mask
        na_mask = df_filled[label_name + '_filled'].isna()

    return df_filled, uniform_subsets
```

```python
trans_df_t = trans_df.copy()
trans_df_t['is_decline0'] = (trans_df_t['totalDeclines'] == 0).astype(int)
trans_df_t['is_decline1'] = (trans_df_t['totalDeclines'] == 1).astype(int)
trans_df_t['is_trans_smaller_2000'] = (trans_df_t['transactionAmount'] < 2000).astype(int)

binary_cols = ['isForeign', 'isTrustedSeller', 'hasHistory', 'is_decline0','is_decline1','is_trans_smaller_2000']
df_filled, uniform_subsets = stepwise_fraud_finding(trans_df_t, 'isFraud', binary_cols)
```

```python
import numpy as np

def rooks_fun(N, m):
    i_l = {}
    positions = [(i, j) for i in range(1, N+1) for j in range(1, N+1)]
    if m > N * N:
        m = N * N  # Limit the number of rooks to the number of available positions
    chosen_positions = np.random.choice(len(positions), m, replace=False)
    for i in range(m):
        pos = positions[chosen_positions[i]]
        i_l[f'{i}'] = pos
    return i_l

def chess_rooks_optimize(N, m):
    if N == 1 and m >= 1 or m >= N * N:
        return 0
    board = np.zeros((N, N))
    rooks = rooks_fun(N, m)
    for id in range(len(rooks)):
        row, col = rooks[f'{id}']
        board[row-1][col-1] = 1  # Place the rook
        for i in range(N):
            if i != row-1:
                board[i][col-1] = 2  # Mark attacked cells in the same column
            if i != col-1:
                board[row-1][i] = 2  # Mark attacked cells in the same row

    unattacked_cells = 0
    for row in range(N):
        for col in range(N):
            if board[row][col] == 0:  
                unattacked_cells += 1  
    return unattacked_cells

def read_files_chess_O(file_path):
    list_res = []
    with open(file_path) as f:
        data_out = f.read().splitlines()
    for line in data_out:
        parts = line.split()
        N = int(parts[0])
        m = int(parts[1])
        print(N, m)
        list_res.append(chess_rooks_optimize(N, m))
    return list_res

def upload_file(output_path, results):
    # This function is a placeholder for uploading the output file
    with open(output_path, 'w') as f:
        for result in results:
            f.write(f"{result}\n")

# Example usage:
# results = read_files_chess_O('input.txt')
# upload_file('output.txt')
```

```python
def Grade_year(school):
    columns = ['Year 3 - Top Scoring', 'Year 4 - Top Scoring', 'Year 5 - Top Scoring', 'Year 6 - Top Scoring']
    columns2 = ['Year 3 - All Students Tested', 'Year 4 - All Students Tested', 'Year 5 - All Students Tested', 'Year 6 - All Students Tested']
    
    a = schools_df[schools_df['School Index'] == school][columns].values
    b = schools_df[schools_df['School Index'] == school][columns2].values
    
    # Filter out zero values
    b = b[a != 0]
    a = a[a != 0]
    if len(a) == 0:
        return 0
    # Calculate the maximum ratio
    ratios = a / b
    max_ratio_index = ratios.argmax()
    
    # Map the index to the corresponding year
    year_mapping = {0: 3, 1: 4, 2: 5, 3: 6}
    max_year = year_mapping[max_ratio_index]
    
    return max_year

   
```

```python
def within_a_range(district,dmin,dmax):
    a = schools_df[schools_df['District'] == district]['Average Proficiency Score'].values
    list_a = a[(dmin <= a) & (a <= dmax)]
    if len(list_a) ==0:
        return 0
    targ_schools = schools_df[(schools_df['District'] == district) & (schools_df['Average Proficiency Score'].isin(list_a))]['School Index'].unique()
    list_schools = [schools_df[schools_df['School Index'] == school]['Student Achievement Rating'] for school in targ_schools]
    return round(np.mean(list_schools),2)

def within_a_range_file(file_path):
    list_res = []
    with open(file_path) as f:
        data_out = f.read().splitlines()
    for line in data_out:
        parts = line.split()
        N = int(parts[0])
        B = int(parts[1])
        A = int(parts[2])
        list_res.append(within_a_range(N, B, A))
    return within_a_range(N, B, A)

def upload_similarity(output_path, results):
    # This function is a placeholder for uploading the output file
    with open(output_path, 'w') as f:
        for result in results:
            f.write(f"{result}\n")


```
